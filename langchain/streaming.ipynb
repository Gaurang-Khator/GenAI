```python
import os
from dotenv import load_dotenv 

load_dotenv()

os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite-preview")

for i in model.stream("Write 200 words about GenAI"):
    print(i.text)
```

```python
responses = model.batch([
    "Write 200 words about GenAI",
    "Write 200 words about Machine Learning",
    "Write 250 words about Deep Learning"
])

for res in responses:
    print(res.text)
```

```python
model.batch([
    "Write 200 words about GenAI",
    "Write 200 words about Machine Learning",
    "Write 250 words about Deep Learning"
],
config={
    "max_concurrency": 2
})
```
